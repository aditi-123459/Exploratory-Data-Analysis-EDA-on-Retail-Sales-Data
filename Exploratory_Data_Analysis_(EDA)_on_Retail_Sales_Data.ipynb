```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading Dataset
df = pd.read_excel("retail_sales_dataset.xlsx")
```

```python

df.head()
```

```python
#Data Understanding
df.shape
```

```python
#Initial Data Exploration
df.info()
```

```python
#Checking Column Names

df.columns
```

```python
#Missing Value Analysis
df.isnull().sum()
```

```python
#Step 5: Filling Missing Values using Forward Fill
df.fillna(method='ffill', inplace=True)
```

```python
# Step 3: Removing Duplicate Rows

df.drop_duplicates(inplace=True)
```

```python
# Step: Converting Date Columns to Datetime Format

df['BirthDate'] = pd.to_datetime(df['BirthDate'])
df['JoinDate'] = pd.to_datetime(df['JoinDate'])
```

```python
# Step: Understanding Dataset Structure

df.info()
```

```python

df.isnull().sum()

```

```python
## Step: Creating Age Feature from BirthDate

today = pd.Timestamp('today')

df['Age'] = (today - df['BirthDate']).dt.days // 365
```

```python
df[['BirthDate', 'Age']].head()
```

```python
## Step: Creating Customer Tenure Feature from JoinDate

df['CustomerTenure_Years'] = (today - df['JoinDate']).dt.days // 365
```

```python
## Step: Verifying Customer Tenure Feature Creation

df[['JoinDate', 'CustomerTenure_Years']].head()
```

```python
# Step: Generating Statistical Summary of Numerical Features

df.describe()
```

```python
## Step: Analyzing Gender Distribution

df['Gender'].value_counts()
```

```python
## Step: Visualizing Gender Distribution

plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title("Customer Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
## Step: Visualizing Age Distribution using Histogram

plt.figure()
plt.hist(df['Age'], bins=10)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Step: Analyzing City-wise Customer Distribution
city_counts = df['City'].value_counts()
city_counts
```

```python
# Step: Viewing Top 10 Cities by Customer Count

city_counts.head(10)
```

```python
## Step: Visualizing Top 10 Cities by Customer Count
plt.figure()
city_counts.head(10).plot(kind='bar')
plt.title("Top 10 Cities by Customer Count")
plt.xlabel("City")
plt.ylabel("Number of Customers")
plt.show()
```

```python
df['JoinYear'] = df['JoinDate'].dt.year
```

```python
## Step: Analyzing Customer Join Year Distribution

join_year_counts = df['JoinYear'].value_counts().sort_index()
join_year_counts
```

```python
## Step: Visualizing Customer Join Trend Over Years

plt.figure()
join_year_counts.plot()
plt.title("Customer Join Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Step: Visualizing Correlation Between Numeric Features

corr = df[['Age', 'CustomerTenure_Years']].corr()
plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()

```
